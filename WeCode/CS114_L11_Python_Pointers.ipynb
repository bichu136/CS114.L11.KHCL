# Đếm node lá
###Mô tả:
**Bài toán**: Cho một dãy số, hãy cho biết số node lá của cây nhị phân tìm kiếm thu được khi thêm lần lượt các số trong dãy vào cây. \\

**INPUT**: Một dãy số nguyên khác 0, mỗi số trên một hàng. Dãy số tận cùng bằng số 0 \\

**OUTPUT**: Số node lá trong cây

```
# class node của cây
class Node:
    # hàm khởi tạo
    def __init__(self, value=None, bool=True):
        self.value = value  # gán giá trị cho node
        self.left = None  # node con bên trái là rỗng
        self.right = None  # node con bên phải là rỗng

    # hàm thay đổi giá trị
    def change_val(self, value):
        self.value = value  # thay giá trị cũ bằng giá trị mới

# hàm thêm node con
def insert(node, child):
    if node is None:
        return
    if child.value == node.value:  # nếu giá trị bị trùng thì loại
        return
    # nếu giá trị node con lớn hơn giá trị node đang xét, thêm node con vào
    # phải
    if child.value > node.value:
        if node.right is not None:
            insert(node.right, child)
        else:
            node.right = child
            return
    # nếu không phải, thêm node con vào bên phải
    else:
        if node.left is not None:
            insert(node.left, child)
        else:
            node.left = child
            return

# hàm đếm node lá
def count_leaf(tree, count):
    if tree is None:
        return
    if tree.left is not None:
        count = count_leaf(tree.left, count)
    if tree.right is not None:
        count = count_leaf(tree.right, count)
    if tree.left is None and tree.right is None:
        count += 1
    return count


get = input()
root = Node()
while get != '0':  # Nếu input là 0 thì dừng
    if root.value is None:
        root.change_val(int(get))
    else:
        new_node = Node(int(get))
        insert(root, new_node)
    get = input()

counter = 0
counter = count_leaf(root, counter)
print(counter)
```

## Chú thích
Mỗi node có các thuộc tính value, left và right
### Thuật toán
1. Với mỗi số của input, kiểm tra nếu khác 0 thì xét, bằng 0 thì dừng
2. Bước xét:
- Nếu root chưa có value: gắn value bằng số đầu tiên của input
- Nếu đã có, thêm số từ input vào cây với root làm node đầu tiên
- Thêm vào tree (hàm insert):
Xây dựng một cây đảm bảo:
  - Các node con bên trái có giá trị nhỏ hơn giá trị node cha
  - Các node con bên phải có giá trị lớn hơn giá trị node cha
  - Các node trên cây không có giá trị trùng nhau
3. Đếm node lá:
  - Đếm tất cả các node không có node con


# Chiều cao cây
### Mô tả:
**Bài toán**: Bạn Bình đã làm xong bài danh sách liên kết giờ chuyển qua bài cây nhị phân tìm kiếm. Hỏi nếu Bình đem danh sách đó nhập vào cây nhị phân tìm kiếm thì cây cao bao nhiêu? \\
**INPUT**: Input gồm nhiều dòng, mỗi dòng sẽ có cấu trúc ở 1 trong 4 dạng sau:

- Dạng 0: Dòng bắt đầu bằng con số 0, theo sau là một số nguyên dương < 1000, chương trình cần phải thêm con số này vào đầu danh sách

- Dạng 1: Dòng này bắt đầu bằng con số 1, theo sau là một số nguyên dương < 1000, chương trình cần phải thêm con số này vào cuối danh sách

- Dạng 2: Dòng này bắt đầu bằng con số 2, theo sau là 2 số nguyên a, b < 1000, chương trình cần tìm vị trí đầu tiên mà số a xuất hiện trong danh sách, sau đó thêm số b vào sau số này. Nếu số a không có trong danh sách, thêm b vào đầu danh sách

- Dạng 3: Dòng này bao gồm duy nhất một con số 3. Đây là dòng cuối cùng trong input, báo hiệu input kết thúc


```
from collections import deque


# class node của cây
class Node:
    # hàm khởi tạo
    def __init__(self, value=None, bool=True):
        self.value = value  # gán giá trị cho node
        self.left = None  # node con bên trái là rỗng
        self.right = None  # node con bên phải là rỗng

    # hàm thay đổi giá trị
    def change_val(self, value):
        self.value = value  # thay giá trị cũ bằng giá trị mới

# hàm thêm node con
def insert(node, child):
    if node is None:
        return
    if child.value == node.value:  # nếu giá trị bị trùng thì loại
        return
    # nếu giá trị node con lớn hơn giá trị node đang xét, thêm node con vào
    # phải
    if child.value > node.value:
        if node.right is not None:
            insert(node.right, child)
        else:
            node.right = child
            return
    # nếu không phải, thêm node con vào bên phải
    else:
        if node.left is not None:
            insert(node.left, child)
        else:
            node.left = child
            return

# hàm tính chiều cao cây            
def depth(tree):
    if tree is None:
        return 0
    left = depth(tree.left)
    right = depth(tree.right)
    if left > right:
        return left + 1
    else:
        return right + 1
        
get = input()
list = deque()
while get != '3':
    get = [int(s) for s in get.split()]
    if get[0] == 0:
        list.appendleft(get[1])
    else:
        if get[0] == 1:
            list.append(get[1])
        else:
            if get[1] in list:
                i = list.index(get[1])
                if i == len(list) - 1:
                    list.append(get[2])
                else:
                    list.insert(i + 1, get[2])
    get = input()

# hàm tạo cây từ mảng
def tree_of_list(list):
    root = Node(None)
    for var in list:
        if root.value is None:
            root.change_val(var)
        else:
            new_node = Node(var)
            insert(root, new_node)
    return root


root = tree_of_list(list)
rank = depth(root)
print(rank)
```

## Chú thích
Sử dụng cấu trúc deque của thư viện collections \\
Lý do: append và pop nhanh hơn so với list

  - Chiều cao của node lá là 0

# Duyệt cây theo chiều rộng
### Mô tả
Cho một dãy số. Lần lượt thêm các số trong dãy vào một cây nhị phân tìm kiếm, sau đó duyệt và xuất cây ra màn hình theo thứ tự duyệt theo chiều rộng (các node có mức thấp được xuất trước, đối với các node có cùng mức xuất node có giá trị nhỏ trước) \\
**INPUT**: Một dãy số nguyên khác 0, mỗi số trên một hàng. Dãy số tận cùng bằng số 0 \\
**OUTPUT**:
Xuất trên một hàng giá trị các node trên cây khi duyệt theo chiều rộng

```
# class node của cây
class Node:
    # hàm khởi tạo
    def __init__(self, value=None, bool=True):
        self.value = value  # gán giá trị cho node
        self.left = None  # node con bên trái là rỗng
        self.right = None  # node con bên phải là rỗng

    # hàm thay đổi giá trị
    def change_val(self, value):
        self.value = value  # thay giá trị cũ bằng giá trị mới

# hàm thêm node con
def insert(node, child):
    if node is None:
        return
    if child.value == node.value:  # nếu giá trị bị trùng thì loại
        return
    # nếu giá trị node con lớn hơn giá trị node đang xét, thêm node con vào
    # phải
    if child.value > node.value:
        if node.right is not None:
            insert(node.right, child)
        else:
            node.right = child
            return
    # nếu không phải, thêm node con vào bên phải
    else:
        if node.left is not None:
            insert(node.left, child)
        else:
            node.left = child
            return

# hàm tính bậc cho từng node trong cây
def sub_visit(tree, dict, rank=0): 
    if tree is None:
        return
    dict[tree.value] = rank
    if tree.left is not None:
        sub_visit(tree.left, dict, rank+1)
    if tree.right is not None:
        sub_visit(tree.right, dict, rank+1)


root = None
get = int(input())
while get != 0:
    if root:
        new = Node(get)
        insert(root, new)
    else:
        root = Node(get)
    get = int(input())

# lưu trữ giá trị của node và bậc của nó
dict = {}
sub_visit(root, dict)
i = 0
while i in dict.values():
    list = []   # Lưu trữ các node có cùng bậc i
    for key, val in dict.items():
        if val == i:
            list.append(key)
    for key in list:
        print(key, end=' ')
    i += 1
```

2. Hàm tính bậc:
  - Bậc của node gốc là 0
  - Bậc node con sẽ bằng bậc node cha cộng 1
  - Thêm vào dictionary với key là giá trị của node và giá trị là bậc của node đó
3. Xuất kết quả
  - Xuất tất các giá trị của node có cùng bậc cho từng bậc tăng dần trong dictionary
